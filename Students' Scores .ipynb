```python
df['Avg'] = df[test_cols].mean(axis=1)
df['Sd'] = df[test_cols].std(axis=1)

fig, ax = plt.subplots(ncols=2, figsize=(10,5))
sns.histplot(data=df, x="Avg", kde=True, ax=ax[0], bins=15)
ax[0].set(title="Means of Test Scores", xlabel="Test Score")

sns.histplot(data=df, x="Sd", kde=True, ax=ax[1], bins=15)
ax[1].set(title="Std Dev of Test Scores", xlabel="Test Score")

plt.show()
```

According to the mean of the test scores, it's safe to say that average test scores among the students are between 60 and 80. 
As for the standard deviation, it's range is from 5.0 to 7.5. 

Achiever


From the means histogram, we see that there are roughly three categories of students;

-those who have mean marks between 30 and 55 (the low achievers)

-those who have mean marks between 55 and 80 (the medium achievers)

-those who have mean marks between 80 and 100 (the high achievers)

```python
def classify_achiever(tl):
    mean = tl.mean()
    if mean < 55:
        return 'Low'
    elif mean >= 55 and mean < 80:
        return 'Medium'
    else:
        return 'High'
```

```python
df['achiever'] = df[test_cols].apply(classify_achiever, axis=1)
```

```python
df.head()
```

According to the data above, we can see the distrubution of the amount of each class of achiever. 

```python
sns.countplot(data=df, x="achiever")
plt.show()
```

According to this chart, we can see that the numbers are relatively even across the board, which "Medium Achiever" being slightly higher than the "High Achiever" and the "Low Achiever". 

```python
test_means = df[test_cols].mean()
ax = sns.lineplot(data=test_means)
ax.set_xlabel("Test # - 1")
ax.set_ylabel("Mean Marks")
ax.set_xticks(range(0, 12))
plt.show()
```

According to this line graph, we can see the scores starting to have a massive drop, but after test 9, we began to see a slight increase in the scores. With this new information, we can tell that there is something that has to be done to help get students to increase their tests scores. 

Overall, there has to be more solutions to help students improve their scores because if this keeps up, then there's a chance that these scores could get worse.
